Queremos evaluar si el nuevo proceso de checkout (Grupo B) tiene un impacto significativo en el uso de cupones respecto al checkout tradicional (Grupo A).

### **Hipótesis**
- $H_0$: $p_B - p_A \leq 0$ (El nuevo proceso no aumenta significativamente el uso de cupones).
- $H_1$: $p_B - p_A > 0$ (El nuevo proceso aumenta significativamente el uso de cupones).

$$
\text{Nivel de significancia: } \alpha = 0.05
$$

---

## **Cálculos**

### **1. Proporciones de éxito**
Para ambos grupos calculamos las proporciones de uso de cupones:

$$
\hat{p}_A = \frac{x_A}{n_A} = \frac{130}{500} = 0.26
$$

$$
\hat{p}_B = \frac{x_B}{n_B} = \frac{175}{500} = 0.35
$$

---

### **2. Proporción combinada ($\hat{p}$)**
Calculamos la proporción combinada, bajo la hipótesis nula ($H_0$):

$$
\hat{p} = \frac{x_A + x_B}{n_A + n_B} = \frac{130 + 175}{500 + 500} = \frac{305}{1000} = 0.305
$$

---

### **3. Error estándar de la diferencia**
La fórmula para el error estándar es:

$$
SE = \sqrt{\hat{p} \cdot (1 - \hat{p}) \cdot \left(\frac{1}{n_A} + \frac{1}{n_B}\right)}
$$

Sustituyendo los valores:

$$
SE = \sqrt{0.305 \cdot (1 - 0.305) \cdot \left(\frac{1}{500} + \frac{1}{500}\right)}
$$

$$
SE = \sqrt{0.305 \cdot 0.695 \cdot 0.004} = \sqrt{0.0008479} \approx 0.0291
$$

---

### **4. Estadístico $z$**
Calculamos el estadístico $z$ con la fórmula:

$$
z = \frac{\hat{p}_B - \hat{p}_A}{SE}
$$

Sustituyendo:

$$
z = \frac{0.35 - 0.26}{0.0291} \approx \frac{0.09}{0.0291} \approx 3.09
$$

---

### **5. Valor crítico ($z_{crit}$)**
Para un nivel de significancia $\alpha = 0.05$ (unilateral), el valor crítico es:

$$
z_{crit} = 1.645
$$

---

## **Decisión**

Comparando los valores:

- $z = 3.09$
- $z_{crit} = 1.645$

Dado que $z > z_{crit}$, rechazamos $H_0$. Esto indica que hay evidencia significativa de que el nuevo proceso de checkout aumenta el uso de cupones.

---

## **Conclusión**

El nuevo proceso de checkout tiene un impacto positivo significativo en el uso de cupones.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Datos
n_A = 500  # Tamaño del grupo A
n_B = 500  # Tamaño del grupo B
x_A = 130  # Éxitos en el grupo A
x_B = 175  # Éxitos en el grupo B

# Proporciones
p_A = x_A / n_A
p_B = x_B / n_B

# Proporción combinada bajo H0
p_combined = (x_A + x_B) / (n_A + n_B)

# Error estándar de la diferencia
SE = np.sqrt(p_combined * (1 - p_combined) * (1 / n_A + 1 / n_B))

# Estadístico z
z_observed = (p_B - p_A) / SE

# Valor crítico para alfa = 0.05 (unilateral)
z_crit = norm.ppf(0.95)

# Resultados
print(f"Proporción A: {p_A:.4f}")
print(f"Proporción B: {p_B:.4f}")
print(f"Proporción combinada: {p_combined:.4f}")
print(f"Error estándar: {SE:.4f}")
print(f"Estadístico z observado: {z_observed:.2f}")
print(f"Valor crítico z: {z_crit:.2f}")

# Gráfica de la distribución normal
x = np.linspace(-4, 4, 500)
pdf = norm.pdf(x)

plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Normal Distribution', color='blue')
plt.fill_between(x, pdf, where=(x >= z_crit), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z_crit, color='red', linestyle='--', label=f'Critical z = {z_crit:.2f}')
plt.axvline(z_observed, color='green', linestyle='--', label=f'Observed z = {z_observed:.2f}')
plt.title('Z-Test: Comparison of Observed and Critical Values', fontsize=14)
plt.xlabel('Z-Score', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.show()





```
